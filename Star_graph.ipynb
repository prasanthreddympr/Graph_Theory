```python
import networkx as nx

from random import uniform, random, choice, sample, randint

import matplotlib.pyplot as plt

import random
```

```python
#create an empty graph with 10 nodes
G=nx.empty_graph(10)
```

```python
for i in range(1,G.number_of_nodes()):
    #add edge between 0 and node i. Here the hub is 0. So every node is connected to node zero once.
    G.add_edge(i,0)
```

```python
nx.draw(G,with_labels=1)
plt.show()
```
