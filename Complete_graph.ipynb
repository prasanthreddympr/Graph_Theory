```python
import networkx as nx

from random import uniform, random, choice, sample, randint

import matplotlib.pyplot as plt

import random

```

```python
#first create an empty graph with 10 nodes
G=nx.empty_graph(10)
```

```python
nx.draw(G,with_labels=1)
plt.show()
```

```python
#coding an complete graph
#traverse the nodes of the graph
for i in G.nodes():
    for j in G.nodes():
        #to avoid self loops we check if i is not same as j
        if i!=j:
            #add the edge
            G.add_edge(i,j)
```

```python
nx.draw(G,with_labels=1)
plt.show()
```
